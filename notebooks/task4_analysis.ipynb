```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("../data/raw/reviews_with_themes.csv")

df.head()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(
    data=df,
    x="bank",
    hue="sentiment_label"
)

plt.title("Sentiment Distribution by Bank")
plt.xlabel("Bank")
plt.ylabel("Number of Reviews")

plt.show()
```

### Insight

BOA shows a relatively higher proportion of negative reviews compared to CBE and Dashen. Negative feedback across all banks is strongly associated with login issues, slow transactions, and app reliability problems.

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x="bank",
    y="rating"
)

plt.title("Rating Distribution by Bank")

plt.show()
```

### Insight

CBE and Dashen generally maintain higher customer ratings, while BOA demonstrates greater rating variability and lower median scores.

```python
theme_counts = df["theme"].value_counts()

plt.figure(figsize=(10,6))

theme_counts.plot(kind="bar")

plt.title("Most Common Customer Complaint Themes")

plt.xlabel("Theme")
plt.ylabel("Frequency")

plt.show()
```

### Insight

The most recurring issues involve transaction delays, login failures, OTP problems, and app stability concerns. These themes appear consistently across multiple banks.

## Ethical Considerations

The analysis is subject to several limitations and potential biases:

- Users are more likely to leave reviews after negative experiences (negativity bias)
- Some reviews may be fake, duplicated, or emotionally exaggerated
- Google Play reviews may not represent the entire customer population
- Mixed-language reviews can reduce NLP accuracy
